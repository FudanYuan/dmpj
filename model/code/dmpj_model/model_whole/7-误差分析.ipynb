### 定义数据处理的类型：1.with_fill（使用填充数据）；2.without_fill（不使用填充数据）

```python
type = 2
```

### 加载真实数据集

```python
# 导入真实数据集
dataset_real = pd.read_csv('./energy_predict_data/data_prehandle/5_data_split/test/test_%s.csv' % utils.getType(type))
# 数据集大小
count = (int)(np.max(dataset_real['vehicle_id']))
# 数据描述
pd.set_option('precision', 2)
dataset_real.describe()
```

### 加载预测数据集

```python
dataset_pred = pd.read_csv('./energy_predict_data/predict_result/data_pred_%s.csv' % utils.getType(type))
# 数据描述
pd.set_option('precision', 2)
dataset_pred.describe()
```

### 算法列表

```python
list_columns = dataset_pred.columns.tolist()[3:]
list_columns
```

### 获取各算法下的误差并保存

```python
score = utils.getScore(dataset_real['charge_energy'], dataset_pred, list_columns)
score.to_csv('./energy_predict_data/predict_result/model_score_%s.csv' % utils.getType(type), index = True)
```
