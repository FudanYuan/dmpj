```python
data_root = '../energy_predict_data'  # 改成自己电脑上的路径即可
dataset = pd.read_csv(data_root + '/predict_data_e_train.csv')
header = dataset.columns.values.tolist()
dataset.head()
```

```python
testA = pd.read_csv(data_root +'/testA.csv')
testA_car = []
testA_car.append(testA[testA['vehicle_id']==1])
testA_car.append(testA[testA['vehicle_id']==2])
testA_car.append(testA[testA['vehicle_id']==3])
testA_car.append(testA[testA['vehicle_id']==4])
testA_car.append(testA[testA['vehicle_id']==5])
testA_car[2]
```

```python
dataset_car = []
dataset_car.append(dataset[dataset["vehicle_id"]==1])
dataset_car.append(dataset[dataset["vehicle_id"]==2])
dataset_car.append(dataset[dataset["vehicle_id"]==3])
dataset_car.append(dataset[dataset["vehicle_id"]==4])
dataset_car.append(dataset[dataset["vehicle_id"]==5])
dataset_car[0].head()
```

```python
from sklearn.cluster import KMeans
perfect_data = pd.read_csv(data_root + '/perfect.csv')
# 获取列名
header = perfect_data.columns.values.tolist()
#聚类和填充需要操作的列名
compute_header = ['charge_start_soc','charge_end_soc','charge_start_U','charge_end_U','charge_start_I','charge_end_I','charge_energy']
cluster_source = perfect_data[compute_header]
cluster_source_car = []
cluster_source_car.append(cluster_source[perfect_data['vehicle_id']==1])
cluster_source_car.append(cluster_source[perfect_data['vehicle_id']==2])
cluster_source_car.append(cluster_source[perfect_data['vehicle_id']==3])
cluster_source_car.append(cluster_source[perfect_data['vehicle_id']==4])
cluster_source_car.append(cluster_source[perfect_data['vehicle_id']==5])
cluster = []
labels = []
for source in cluster_source_car:
    kmeans = KMeans(n_clusters=5, random_state=0).fit(source[compute_header])
    cluster.append(kmeans.cluster_centers_)
    labels.append(kmeans.labels_)
cluster[0]
```

```python
def distance(vectorA=[], vectorB=[]):
    result = 0.0
    for i in range(len(vectorB)):
        x=vectorA[i]
        y=vectorB[i]
        if (not pd.isnull(y))and (y!=0.0):  
            result += (x-y)*(x-y)
    return np.sqrt(result)
```

```python
def filling(vectorA=[], vectorB=[]):
    for i in range(len(vectorB)):
        x = vectorA[i]
        y = vectorB[i]
        if pd.isnull(y) or (y==0.0):
            vectorB[i] = x
    return vectorB
```

```python
testA_car_result = []
compute_test_header = ['charge_start_soc','charge_end_soc','charge_start_U','charge_end_U','charge_start_I','charge_end_I']
for i in range(len(testA_car)):
    data_source = testA_car[i]
    tmp_data = data_source[compute_test_header]
    tmp_data = tmp_data.values
    for j in range(len(tmp_data)):
        maxD = float('inf')
        maxV = []
        data = tmp_data[j]
        for center in cluster[i]:
            dis = distance(center,data)
            if dis < maxD:
                maxD = dis
                maxV = center
        tmp_data[j] = filling(maxV,data)
    tmp_data = pd.DataFrame(tmp_data,columns=compute_test_header)
    for col in compute_test_header:
        for i in range(len(data_source[col])):
            data_source[col][i] = tmp_data[col][i]
    testA_car_result.append(data_source)
testA_car = testA_car_result
testA_car[2]
```

```python
import copy
header_v1=copy.deepcopy(header)  # 深拷贝
#header_v1.remove('vehicle_id')
header_v1.insert(4, 'delta_mileage')
for i in range(len(testA_car)):
    tmp_test = testA_car[i]
    tmp_data = dataset_car[i]
    tmp_test['uuid']=1
    tmp_data['uuid']=0
    tmp_data = tmp_data.append(tmp_test,sort=False)
    tmp_data.sort_values('mileage',inplace=True)
    mileage_list = tmp_data['mileage'].tolist()
    # 增加里程充电前汽车行驶里程数
    delta_mileage_list = [0] * len(mileage_list)
    for j in range(1, len(mileage_list)):
        delta_mileage = mileage_list[j] - mileage_list[j-1]
        delta_mileage_list[j] = delta_mileage
    tmp_data['delta_mileage'] = delta_mileage_list
    tmp_data = tmp_data[tmp_data['uuid']==1]
    testA_car[i] = tmp_data[header_v1]
testA_car[2]
```

```python
import time
def str2time(string):
    year = string[0:4]
    month = string[4:6]
    day = string[6:8]
    hour = string[8:10]
    minute = string[10:12]
    second = string[12:14]
    time_str = '%s-%s-%s %s:%s:%s' % (year, month, day, hour, minute, second)
    timestamp = time.mktime(time.strptime(time_str, '%Y-%m-%d %H:%M:%S'))  # 变为时间戳
    return int(timestamp)

# 将矩阵内的字符串转化为时间戳
def calVecTimeStamp(vector):
    timestamps = []
    for v in vector:
        timestamp = str2time(str(v))
        timestamps.append(timestamp)
    return timestamps
```

```python
header_v1.insert(3, 'charge_duration')
testA_car_r = []
for tmp_test in testA_car:
    testA_charge_end_time = np.array(calVecTimeStamp(tmp_test['charge_end_time']))
    testA_charge_start_time = np.array(calVecTimeStamp(tmp_test['charge_start_time']))
    tmp_test['charge_duration'] = (testA_charge_end_time - testA_charge_start_time).tolist()
    tmp_test = tmp_test[header_v1]
    testA_car_r.append(tmp_test)
testA_car = testA_car_r
testA_car[0]
```
